```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm

# Load the data
your_data = pd.read_csv("Social_Network_Ads.csv")
your_data['Purchased'] = pd.Categorical(your_data['Purchased'])

# Set up the figure and axis for the boxplot
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Boxplot
sns.boxplot(x='Purchased', y='EstimatedSalary', data=your_data, palette=["skyblue", "lightgreen", "lightcoral"], ax=axes[0])
axes[0].set_title("Comparison of Expected Salary by Purchased Behavior with Outliers")
axes[0].set_xlabel("Purchased Behavior")
axes[0].set_ylabel("Expected Salary")

# Histogram with Normal Curve Overlay
sns.histplot(your_data['EstimatedSalary'], bins=30, kde=True, color="skyblue", ax=axes[1])
sns.lineplot(x=np.linspace(0, 160000, 1000), y=norm.pdf(np.linspace(0, 160000, 1000), loc=np.mean(your_data['EstimatedSalary']), scale=np.std(your_data['EstimatedSalary'])), color="red", ax=axes[1])
axes[1].set_title("Histogram With Normal Curve Overlay")
axes[1].set_xlabel("Expected Salary")
axes[1].set_ylabel("Density")

# Show the plots
plt.tight_layout()
plt.show()
```
